```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load predictions
df = pd.read_csv("../Data/processed/player_test_predictions.csv")
```

# 1. Line plot: Actual vs Predicted Points

```python
plt.figure(figsize=(12,6))
plt.plot(df["Date"], df["Points_actual"], label="Actual Points", marker="o")
plt.plot(df["Date"], df["Points_pred"], label="Predicted Points", marker="x")
plt.xticks(rotation=45)
plt.ylabel("Points")
plt.title("Actual vs Predicted Points")
plt.legend()
plt.tight_layout()
plt.show()
```

# 2. Scatter plot with line of perfect fit

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x="Points_actual", y="Points_pred", data=df)
plt.plot([df["Points_actual"].min(), df["Points_actual"].max()],
         [df["Points_actual"].min(), df["Points_actual"].max()],
         "r--")
plt.xlabel("Actual Points")
plt.ylabel("Predicted Points")
plt.title("Scatter: Actual vs Predicted Points")
plt.show()
```

# 3. Probability calibration histogram for Points line

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Points_over_prob_at_29.5"], bins=10, kde=False)
plt.xlabel("Predicted Probability of Over 29.5")
plt.ylabel("Frequency")
plt.title("Distribution of Predicted Probabilities (Points)")
plt.show()
```
